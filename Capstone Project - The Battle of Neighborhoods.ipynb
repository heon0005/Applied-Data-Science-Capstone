## Foursquare API Data:¶

We will need data on different locations in different neighborhoods in that particular autonomous region. Use the "Foursquare" location information to get that information. Foursquare is a location data provider that provides information about all types of venues and events in your area of interest. Such information includes venue names, locations, menus, and even photos. As a result, the Four Square Location Platform is used as the sole data source because all required information provided can be obtained via the API.

Once we found the list of neighborhoods, connect to the Foursquare API to collect information about venues in all neighborhoods. For each neighborhood, we chose a radius of 100 meters.

The data obtained from Foursquare contained information about venues within a specified distance from the longitude and latitude of the zip code. The information obtained for each venue is as follows.

1. Neighborhood
2. Neighborhood Latitude
3. Neighborhood Longitude
4. Venue
5. Name of the venue e.g. the name of a store or restaurant
6. Venue Latitude
7. Venue Longitude
8. Venue Category
