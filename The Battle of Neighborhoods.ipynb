# The Battle of Neighborhoods - Business Proposal - Introduction

## Introduction:

The purpose of this project is to help people explore better facilities in their neighborhood. This helps people make wise and efficient decisions in choosing a great district from many other districts in Scarborough, Toronto.

Many people have emigrated to different states in Canada and needed a lot of research for good home prices and a reputable school for their children. This project is for people looking for a better area. Easy access to cafes, schools, supermarkets, medical stores, grocery stores, malls, theaters, hospitals, like-minded people and more.

The project aims to create a feature analysis for people migrating to Scarborough to find the best neighborhood as a comparative analysis between neighborhoods. Features include median home prices and better schools, crime rates in that particular area, road connectivity, weather conditions, proper management of emergencies, and freshwater carried by sewers and recreational facilities. Includes both water resources and excrement of wastewater.

It will help people to get awareness of the area and neighborhood before moving to a new city, state, country or place for their work or to start a new fresh life.
